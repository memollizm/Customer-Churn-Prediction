````python
import numpy as np
import pandas as pd
import tensorflow as tf
````

#### Veri Ön İşleme

````python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values #3. kolondan başlayarak exited kolonuna kadar
Y = dataset.iloc[:, -1].values # Exited kolonu
````

````python
X
````

````python
Y
````

#####  Encoding

* Cinsiyet için Label Encoding

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
````

````python
print(X)  #Female = 0, Male = 1
````

* Ülke kolonu için One Hot Encoding

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
````

````python
print(X) #Fransa=100, İspanya=001, Almanya=010 olarak kodlandı
````

* Verinin test ve eğitim olarak bölünmesi

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
````

* Özellik Ölçeklemesi


> Derin Öğrenme için yapılması zorunludur.



````python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
````

# ANN'yi İnşa Etme

* Sıralı yapay sinir ağı

````python
ann = tf.keras.models.Sequential()
````

* İlk gizli katmanın(Hidden Layer) eklenmesi

>Kaç tane gizli nörona sahip olmak istiyorsak "units=" parametresinde belirtmemiz gerekiyor.


>Aktivasyon fonksiyonuna ("activation=")
karşılık gelen değer hariç tümü için aynı değerler korunur.




````python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
````

* İkinci gizli katmanın eklenmesi


>İkinci gizli katmanın eklenmesi, ilk gizli katmanın eklenmesi ile aynıdır.
  >> !! Eğer nöron sayısı sabit olacaksa


````python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
````

* Çıktı katmanının (Output Layer) eklenmesi

````python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
````

# Training the ANN

* ANN'yi Derleme


> İkili bir sonucu tahmin ederken her zaman `loss=binary_crossentropy` olmalıdır.



````python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

* ANN'nin eğitim seti üzerinde eğitilmesi

````python
ann.fit(X_train, y_train, batch_size=32, epochs=20)
````

# SONUÇLAR
* "accuracy" metriğine göre model %86 doğruluk değerine sahiptir.
* Parametreler değiştirilerek/düzeltilerek doğruluk oranı artırabilir.

# **Örnek**

ANN'yi kullanarak aşağıdaki özelliklere sahip olan müşterinin bankadan ayrılıp ayrılmayacağını tahmin et.

* Geography: France
* Credit Score: 600
* Gender: Male
* Age: 40 years old
* Tenure: 3 years
* Balance: $ 60000

* Number of products: 2
* Kredi kartına sahip mi?: Yes
* Aktif bir üye mi?: Yes
* Estimated Salary: $ 50000


````python
print(ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5)
````

* ANN'yi ölçeklendirdiğimiz için ```sc.transform``` kullanılmalıdır.



* Label Encoding aşamasıdan Fransa: [1,0,0] olarak kodlanmıştı

* Label Encoding aşamasında Male: 1 olarak kodlanmıştı

* Kredi kartına sahip mi? Label Encoding aşamasında 1 olarak kodlanmıştı

* Aktif bir üye mi? Label Encoding aşamasında 1 olarak kodlanmıştı

------------------------------------------

Tahmin ve Test Seti Sonuçları
* Sol sütun yapay sinir ağının tahminleri
* Sağ sütun gerçek sonuçlar

````python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test), 1)), 1))
````

Confusion Matrix

````python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
````
